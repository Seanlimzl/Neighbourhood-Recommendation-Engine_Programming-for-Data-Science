```python
# Data processing
import numpy as np
import pandas as pd
import geopandas as gpd
import pathlib as path

# Visualisation
import matplotlib
from ipyleaflet import Map, GeoData, basemaps, LayersControl
import seaborn
import folium
import mapclassify
```

```python
# Reading CSV as Pandas dataframe
hawkercentres = pd.read_csv(r'Data Sources\1_data\hawker-centres\hawker-centres-2.csv')
hawkercentres.head()
```

# Geodataframe


```python
# Converting hawkercentres to a GeoDataFrame
hawkercentres_gdf = gpd.GeoDataFrame(
    hawkercentres, geometry=gpd.points_from_xy(hawkercentres.longitude, hawkercentres.latitude), crs='WGS84'
)

# This line of code sets the CRS to EPSG3414, the Projected Coordinate System for Singapore.
# Compare these coordinates to the coordinates for the original hawkercentres dataframe.
# The coordinates should be in a 5000-50000 range for longitude and 20000-50000 range for latitude (roughly).
# When you get something like the coordinates in the original hawkercentres dataframe (1.something, 100.something),
# that is in the WGS84 standard format and should be converted like in the following line.
hawkercentres_EPSG3414 = hawkercentres_gdf.to_crs(epsg=3414)

# Save hawkercentres_EPSG3414 to a file
hawkercentres_EPSG3414.to_file('Generated Files/hawkercentres_EPSG3414.shp')
hawkercentres_EPSG3414.head()
```
